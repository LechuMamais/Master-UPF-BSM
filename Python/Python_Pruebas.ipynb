### Ejercicio 4


La ecuaci√≥n cuadr√°tica es una ecuaci√≥n polin√≥mica de segundo grado que tiene la forma general:

`ax^2+bx+c=0`

Donde:

`a`, `b`, y `c` son coeficientes constantes, y `x` es la incognita o la variable que queremos encontrar.

Para resolver una ecuaci√≥n cuadr√°tica, usamos la f√≥rmula cuadr√°tica (ver celda siguiente), donde:


- `a`, `b`, y `c` son los coeficientes de la ecuaci√≥n cuadr√°tica (ax^2 + bx + c = 0).
- El discriminante es Delta = b^2 - 4ac.
  - Si Delta > 0, hay dos soluciones reales.
  - Si Delta = 0, hay una soluci√≥n real doble.
  - Si Delta < 0, las soluciones son complejas.



Resuelve la equaci√≥n de segundo grado, considerando diferentes valores de a, b y c. Muestra un caso con 2 soluciones reales, un caso con una soluci√≥n real doble y un caso con soluciones complejas.

En caso de que el resultado sea un n√∫mero complejo, como accedemos a la parte imaginaria y la parte real?

```python
# Cargamos la librer√≠a display para mostrar f√≥rmulas matem√°ticas
from IPython.display import display, Math

# Mostramos la f√≥rmula cuadr√°tica
display(Math(r'x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}'))
```

```python
def resolver_ecuacion_cuadratica(a, b, c):
  print(f"Ecuaci√≥n: {a}x¬≤ + {b}x + {c} = 0")

  d = b**2 - 4*a*c
  print(f"Discriminante (Œî): {d}")

  if d > 0:
    # Dos soluciones reales
    x1 = (-b + d**0.5) / (2*a)
    x2 = (-b - d**0.5) / (2*a)
    print("üëâ Dos soluciones reales:")
    print(f"x1 = {x1}")
    print(f"x2 = {x2}")
    return x1, x2

  if d == 0:
    # Una soluci√≥n real doble
    x = -b / (2*a)
    print("üëâ Una soluci√≥n real doble:")
    print(f"x = {x}")
    return x

  if d < 0:
    # Soluciones complejas
    parte_real = -b / (2*a)
    parte_imaginaria = d**0.5 / (2*a)
    print("üëâ Soluciones complejas:")

    # Si quieres acceder a la parte real e imaginaria:
    print("Parte real:", parte_real)
    print("Parte imaginaria:", parte_imaginaria)
    return parte_real, parte_imaginaria


# Caso 1: Dos soluciones reales
resolver_ecuacion_cuadratica(1, -3, 2)

# Caso 2: Una soluci√≥n real doble
resolver_ecuacion_cuadratica(1, -2, 1)

# Caso 3: Soluciones complejas
resolver_ecuacion_cuadratica(1, 2, 5)
```

```python
def elementos_pares(lista):
  """
  Funci√≥n que devuelve los elementos pares de una lista
  """
  return [x for x in lista if x % 2 == 0]

lista_2 = ["a", "b", "c", "d", "e", "f"]


def elementos_pares_v2(lista):
  return [x for i, x in enumerate(lista, 1) if i % 2 == 0]


print(list(elementos_pares(list(range(1, 16)))))
print(list(elementos_pares_v2(list(range(1, 16)))))
print(list(elementos_pares_v2(lista_2)))
```

```python
# Respuesta
import unicodedata

def normalizar_texto(texto: str) -> str:
  """Elimina tildes y convierte a min√∫sculas."""
  texto = texto.lower()
  texto = unicodedata.normalize('NFKD', texto).encode('ascii', 'ignore').decode('utf-8')
  return texto


def tramo_electricidad(dia: str, hora: int) -> str:
  """
  Esta funci√≥n recibe el d√≠a de la semana y la hora.
  Devuelve el tramo de electricidad correspondiente a ese d√≠a y hora (string).
  Muestra un mensaje en relaci√≥n al tramo donde nos encontramos.
  Internamente, comprueba que el d√≠a y la hora introducidos sean v√°lidos.
  Si no lo son, lanza el error.
  """

  try:
    # Validar tipos
    if not isinstance(dia, str):
      raise TypeError("[ERROR] El d√≠a debe ser texto (str).")
    if not isinstance(hora, int):
      raise TypeError("[ERROR] La hora debe ser un n√∫mero entero (int).")

    # Validar rango de hora
    if not (0 <= hora < 24):
      print(f"{hora} no es una hora del d√≠a. Introduce un valor entre 0 y 23.")
      return

    dias_finde = ("sabado", "domingo")
    dias_semana = ("lunes", "martes", "miercoles", "jueves", "viernes")

    dia_norm = normalizar_texto(dia)
    if dia_norm not in dias_finde + dias_semana:
      print(f"{dia} no es un d√≠a de la semana. Introduce un valor v√°lido.")
      return

    is_finde = dia_norm in dias_finde

    if is_finde:
      tramo = "valle"
    elif (10 <= hora < 14) or (18 <= hora < 22):
      tramo = "punta"
    elif (8 <= hora < 10) or (14 <= hora < 18) or (hora >= 22):
      tramo = "llano"
    else:
      tramo = "valle"

    mensajes = {
      "punta": "es cuando la electricidad es m√°s cara. Que no se te ocurra encender el AC. Y menos lavar la ropa!",
      "llano": "tiene un precio medio. Si prendes el AC ponlo en 22¬∫C, y esperate para lavar la ropa.",
      "valle": "es cuando la electricidad es m√°s barata. A poner la lavadora!"
    }

    mensaje_final = f"{dia.capitalize()} a las {hora}hs. Estamos en el tramo {tramo}, {mensajes[tramo]}"
    print(mensaje_final)
    return tramo

  except Exception as e:
    print(e)
    return None

# Pruebas
tramo_electricidad('lunes', 13)
tramo_electricidad('sabado', 13)
tramo_electricidad('Miercoles', 23)
tramo_electricidad('luneos', 10)
tramo_electricidad('jueves', 100)
tramo_electricidad(123, 10)
tramo_electricidad('martes', 'diez')

```
