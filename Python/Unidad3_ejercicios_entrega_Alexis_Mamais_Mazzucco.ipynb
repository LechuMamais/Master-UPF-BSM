b) Crea una función que, dado un diccionario que contiene libros y sus respectivas páginas, nos devuelva una lista con los libros que se pueden leer en un número concreto de días. (**1 punto**)

- Como input de la función se deberán indicar el diccionario, el número de páginas que se leen en un día, y el número de días disponibles para leer.
- Como output se deberá devolver la lista con los libros que cumplen la condición, y también se deberá mostrar por pantalla un mensaje, por ejemplo:

`En 10 días puedo leer los libros siguientes: ['Harry Potter 1', 'Orgullo y prejuicio', 'El viento conoce mi nombre', 'El principito']`


(**2 puntos**)

```python
# Respuesta
# a)
def acronimo(frase):
  """
  Función que devuelve el acrónimo de una frase
  """
  return "".join([x[0] for x in frase.split(' ')]).upper()

# b)
def books_readable_in_days(books_dict: dict, pages_per_day: int, days: int) -> list:
  """
  Función que recibe un diccionario de libros, las páginas que se pueden leer por día y la cantidad de días,
  y devuelve una lista con los títulos de los libros que se pueden leer en esos días.
  He tipado los parámetros y el valor de retorno de la función para mayor claridad y robustez.
  """
  books_readable = [book_title for book_title in books_dict if books_dict[book_title] <= pages_per_day * days]
  print(f" En {days} días puedo leer los siguientes libros: {books_readable}")
  return books_readable
```
