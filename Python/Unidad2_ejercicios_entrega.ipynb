### Ejercicio 4


La ecuaci칩n cuadr치tica es una ecuaci칩n polin칩mica de segundo grado que tiene la forma general:

`ax^2+bx+c=0`

Donde:

`a`, `b`, y `c` son coeficientes constantes, y `x` es la incognita o la variable que queremos encontrar.

Para resolver una ecuaci칩n cuadr치tica, usamos la f칩rmula cuadr치tica (ver celda siguiente), donde:


- `a`, `b`, y `c` son los coeficientes de la ecuaci칩n cuadr치tica (ax^2 + bx + c = 0).
- El discriminante es Delta = b^2 - 4ac.
  - Si Delta > 0, hay dos soluciones reales.
  - Si Delta = 0, hay una soluci칩n real doble.
  - Si Delta < 0, las soluciones son complejas.



Resuelve la equaci칩n de segundo grado, considerando diferentes valores de a, b y c. Muestra un caso con 2 soluciones reales, un caso con una soluci칩n real doble y un caso con soluciones complejas.

En caso de que el resultado sea un n칰mero complejo, como accedemos a la parte imaginaria y la parte real?

**(1.5 puntos)**

```python
# Cargamos la librer칤a display para mostrar f칩rmulas matem치ticas
from IPython.display import display, Math

# Mostramos la f칩rmula cuadr치tica
display(Math(r'x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}'))
```

```python
# Respuesta
def resolver_ecuacion_cuadratica(a, b, c):
  """
  Esta funci칩n resuelve la ecuaci칩n cuadr치tica ax + bx + c = 0 y devuelve las soluciones.
  1. Calcula el discriminante (풊 = b - 4ac).
  2. Dependiendo del valor del discriminante, calcula las soluciones:
     - Si 풊 > 0: dos soluciones reales.
     - Si 풊 = 0: una soluci칩n real doble.
     - Si 풊 < 0: soluciones complejas.
  3. Imprime las soluciones.
  """
  print(f"Ecuaci칩n: {a}x + {b}x + {c} = 0")

  d = b**2 - 4*a*c
  print(f"Discriminante (풊): {d}")

  if d > 0:
    print("Dos soluciones reales:")
    x1 = (-b + d**0.5) / (2*a)
    x2 = (-b - d**0.5) / (2*a)
    print(f"x1 = {x1}")
    print(f"x2 = {x2}")
    return x1, x2
  
  if d == 0:
    print("Soluci칩n real doble:")
    x = -b / (2*a)
    print(f"x = {x}")
    return x

  if d < 0:
    print("Soluciones complejas:")
    parte_real = -b / (2*a)
    parte_imaginaria = d**0.5 / (2*a)
    print("Parte real:", parte_real)
    print("Parte imaginaria:", parte_imaginaria)
    return parte_real, parte_imaginaria
  

# Caso 1: Dos soluciones reales
resolver_ecuacion_cuadratica(1, -3, 2)

# Caso 2: Una soluci칩n real doble
resolver_ecuacion_cuadratica(1, -2, 1)

# Caso 3: Soluciones complejas
resolver_ecuacion_cuadratica(1, 2, 5)
```
